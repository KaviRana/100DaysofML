# What is Parts of Speech Tagging 
Part-of-Speech (POS) tagging, also known as grammatical tagging or word-category disambiguation, is a natural language processing (NLP) technique that involves assigning a specific part-of-speech tag to each word in a sentence. The goal of POS tagging is to identify the syntactic category or grammatical role of each word, such as nouns, verbs, adjectives, adverbs, pronouns, prepositions, conjunctions, and more.

# Spacy 
spaCy is an open-source natural language processing (NLP) library written in Python. It is designed to be efficient, fast, and user-friendly, making it popular among researchers, developers, and data scientists working on NLP tasks. spaCy provides pre-trained models for various languages, allowing users to perform a wide range of NLP tasks with ease.

```python
import spacy
```

```python
# Load the English model
nlp = spacy.load('en_core_web_sm')
```

```python
corpus = "I left my room. My room is on the left corridor. In my room , you will find the cupboard in the left. "
doc = nlp(corpus)

for token in doc:
    print(token.text, token.pos_)
```
